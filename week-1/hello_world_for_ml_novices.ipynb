## DataFrames & Series: The Building Blocks of Pandas

Pandas has two main data structures that you'll use constantly:

- **Series**: A one-dimensional array-like object (like a single column of data)
- **DataFrame**: A two-dimensional table with rows and columns (like a spreadsheet)

Let's explore these with our Iris dataset!

````python
# Let's examine our DataFrame
print("DataFrame shape (rows, columns):", tips.shape)
print("DataFrame columns:", list(tips.columns))
````

````python
print("DataFrame info:")
tips.info()
````

## Reading & Writing Data

Pandas can read from and write to many file formats. The most common is CSV, but you can also work with Excel, JSON, SQL databases, and more.

Common read functions:
- `pd.read_csv()` - for CSV files
- `pd.read_excel()` - for Excel files
- `pd.read_json()` - for JSON files

Common write functions:
- `df.to_csv()` - write to CSV
- `df.to_excel()` - write to Excel
- `df.to_json()` - write to JSON

````python
# Reading data with options
iris_data = pd.read_csv('Iris.csv', index_col=0)  # Use first column as index
print("Data with custom index (note, that the index is now the first column):")
iris_data
````

````python
# Writing data to a new CSV file
iris_sample = tips.head(10)  # Take first 10 rows
iris_sample.to_csv('iris_sample.csv', index=False)  # Save without the index
print("\nSaved first 10 rows to 'iris_sample.csv'")

# Check what we just saved
sample_check = pd.read_csv('iris_sample.csv')
print("Verification - saved data shape:", sample_check.shape)
````

Note, you can save the dataframe with the index as well, by omitting `index=False` parameter.   
That way the index will be saved as a separate column in the CSV file. Try it out and see the difference!

## Data Selection & Indexing

Selecting data is one of the most important skills in pandas. Here are the key methods:

- **Column selection**: `df['column_name']` or `df[['col1', 'col2']]`
- **Label-based indexing**: `df.loc[row_label, column_label]`
- **Position-based indexing**: `df.iloc[row_position, column_position]`
- **Boolean indexing**: `df[df['column'] > value]`

Let's practice these techniques!

````python
# Column selection
print("Single column selection:")
print(tips['Species'].head()) # Display first 5 values of 'Species' column
````

````python
print("\nMultiple column selection:")
print(tips[['SepalLengthCm', 'Species']].head()) # Display first 5 rows of selected columns
````

````python
# Label-based indexing (.loc)
print("\nLabel-based selection (first row, Species column):")
print(tips.loc[0, 'Species'])
````

````python
# Position-based indexing (.iloc)
print("\nPosition-based selection (first row, second column):")
print(tips.iloc[0, 1])
````

````python
# Boolean indexing
print("\nBoolean indexing (rows where SepalLengthCm > 5.0):")
long_sepals = tips[tips['SepalLengthCm'] > 5.0]
print(long_sepals[['SepalLengthCm', 'Species']].head())
````

````python
# Combining selections
print("\nCombining selections (specific rows and columns):")
print(tips.loc[0:2, ['SepalLengthCm', 'SepalWidthCm', 'Species']])
````

Note, that `.loc` includes the stop index `2`, while `.iloc` excludes it.  
So, `.iloc` is consistent with Python slicing, but `.loc` is not.  
This is done to make `.loc` more intuitive when working with labeled data.

````python
print("\nEquivalent selection using .iloc:")
print(tips.iloc[0:3, [1, 2, 5]])  # Rows 0 to 2, Columns 1, 2, 5
````
